```python
import numpy as np

# linespace()함수로 넘파이 배열 생성
x = np.linspace(0, 10, 100)
y1 = x
y2 = 2 * x
y3 = 0.5 * x
y4 = 3 * x

# 그래프 그리기
plt.plot(x, y1, label = 'y = x')
plt.plot(x, y2, label = 'y = 2x')
plt.plot(x, y3, label = 'y = 0.5x')
plt.plot(x, y4, label = 'y = 3x')

# 범례 설정(ncol을 2로 설정하여 한 줄에 2개 표시)
plt.legend(loc = 'upper center', ncol = 2)

# 타이틀 및 축 라벨 설정
plt.title('simple Linear Plot Exemple')
plt.xlabel('x')
plt.ylabel('y')

# 그래프 표시
plt.show()
#
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(1, 10, 100)
y = np.exp(x)

# 선형 스케일
plt.subplot(2, 2, 1)
plt.plot(x,y)
plt.title('Linear Scale')

# 로그 스케일
plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.yscale('log')
plt.title('Logarithmic Scale(y-axis)')

# 로그-선형 스케일
plt.subplot(2, 2, 3)
plt.plot(x,y)
plt.xscale('log')
plt.title('Log-Linear Scale(x-axis)')

# 선형-로그 스케일
plt.subplot(2, 2, 4)
plt.plot(x,y)
plt.xscale('log')
plt.yscale('linear')
plt.title('Linear-Logaritmic Scale')

plt.tight_layout() # 서브 그래프 간격 조절
plt.show()
```

```python
import pandas as pd

titanic= pd.read_csv('titanic_selena.csv')
# 객실 등급에 따른 생존자와 사망자의 평균 계산
pclass_survived_mean =  titanic.groupby('Pclass')['Survived'].mean().reset_index()
print(pclass_survived_mean)
```

```python
import matplotlib.pyplot as plt

# 선 그래프 그리기
plt.plot(pclass_survived_mean['Pclass'], pclass_survived_mean['Survived'], marker='o', linestyle='-', color='violet')
plt.title('Survived Rate Variation Across Passenger Classes')
plt.xlabel('Pclass')
plt.ylabel('Survived Rate')
plt.xticks([1,2,3])
plt.grid(True)
plt.show()
```

```python
import pandas as pd

titanic = pd.read_csv('titanic_selena.csv')
# 승선 항구에 따른 생존자의 수 계산
survived_counts =  titanic[titanic['Survived']==1]['Embarked'].value_counts()
survived_counts
```

```python
import pandas as pd
titanic = pd.read_csv('titanic_selena.csv')

# 결측치 처리
titanic = titanic.dropna(subset=['Age', 'Fare'])
# 나이대별 생존자와 사망자 수 계산
# 범주형 분수로 변환
age_groups = pd.cut(titanic['Age'], bins=range(0, 81, 5))
# Age, Survived 기준으로 그룹화
survived_counts= titanic.groupby([age_groups, 'Survived'], observed=False).size().unstack().fillna(0)
survived_counts
```

```python
import matplotlib.pyplot as plt
# 영역 채우기 그래프 그리기
plt.figure(figsize=(10,6))
plt.fill_between(survived_counts.index.astype(str), survived_counts[1], color='purple', alpha=0.9, label='Survived') # 나이대별 생존자
plt.fill_between(survived_counts.index.astype(str), survived_counts[0], color='hotpink', alpha=0.6, label='Not Survived') # 나이대별 사망자
plt.title('Survival Count by Age Group on the Titanic')
plt.xlabel('Age')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```
