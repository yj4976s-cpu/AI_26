```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
# xlabel()함수에 'X-Label' 입력하여 x축에 대한 레이블 표시
plt.xlabel('X-Label')
# # ylabel()함수에 'Y-Label' 입력하여 y축에 대한 레이블 표시
plt.ylabel('y-Label')
plt.show()
```

```python
import matplotlib.pyplot as plt

# plot() 함수의 label 매개변수에 'Square'문자열 입력
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label = 'Square')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')

# legend() 함수를 사용해서 그래프에 범례표시
plt.legend()
plt.show()
```

```python
import numpy as np

# linespace()함수로 넘파이 배열 생성
x = np.linspace(0, 10, 100)
y1 = x
y2 = 2 * x
y3 = 0.5 * x
y4 = 3 * x

# 그래프 그리기
plt.plot(x, y1, label = 'y = x')
plt.plot(x, y2, label = 'y = 2x')
plt.plot(x, y3, label = 'y = 0.5x')
plt.plot(x, y4, label = 'y = 3x')

# 범례 설정(ncol을 2로 설정하여 한 줄에 2개 표시)
plt.legend(loc = 'upper center', ncol = 2)

# 타이틀 및 축 라벨 설정
plt.title('simple Linear Plot Exemple')
plt.xlabel('x')
plt.ylabel('y')

# 그래프 표시
plt.show()
#
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(1, 10, 100)
y = np.exp(x)

# 선형 스케일
plt.subplot(2, 2, 1)
plt.plot(x,y)
plt.title('Linear Scale')

# 로그 스케일
plt.subplot(2, 2, 2)
plt.plot(x,y)
plt.yscale('log')
plt.title('Logarithmic Scale(y-axis)')

# 로그-선형 스케일
plt.subplot(2, 2, 3)
plt.plot(x,y)
plt.xscale('log')
plt.title('Log-Linear Scale(x-axis)')

# 선형-로그 스케일
plt.subplot(2, 2, 4)
plt.plot(x,y)
plt.xscale('log')
plt.yscale('linear')
plt.title('Linear-Logaritmic Scale')

plt.tight_layout() # 서브 그래프 간격 조절
plt.show()
```

```python
from matplotlib import lines
import matplotlib.pyplot as plt
# plot()함수의 세번째 인수값으로 간략화한 문자열을 입력하여 선 종류 설정
plt.plot([1 , 2, 3], [4,4,4], '-', label='solid')
plt.plot([1 , 2, 3], [3,3,3], '--', label='dashed')

# plot()함수의 linestyle 값으로 문자열을 입력하여 선종류 설정
plt.plot([1,2,3], [2,2,2], linestyle = 'dotted', label = 'dotted')
plt.plot([1,2,3], [1,1,1], linestyle = 'dashdot', label = 'dash-dot')

# 축범위와 범례위치 설정
plt.axis([0.8, 3.2, 0.5, 5.0])

plt.legend(loc='upper center', ncol=4)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([4,5,6],"b")    # 'b' blue : 파란색 마커
plt.plot([3,4,5],"ro")   # 'ro' red+circle : 빨간색 + 원형마커
plt.plot([2,3,4],"s")   # 's' square : 사각형 마커
plt.plot([1,2,3],"D")   # 'D' diamond : 마름모 마커
plt.plot([0, 1, 2], marker = '$A$') # '$문자$" : 사용자 지정 문자마커
plt.show()
```
