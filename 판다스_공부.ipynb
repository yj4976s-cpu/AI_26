```python
import pandas as pd
netflix = pd.read_csv('netflix_selena.csv')
# loc[] : 데이터프레임의 행과 열의 이름을 사용
# 전체 행과 title과 release_year의 열을 출력
title_year = netflix.loc[ : , ['title', 'release_year']]
title_year.head()
```

```python
# .loc[] : 데이터프레임의 행과 열의 이름을 사용
# 0번째,1번째 행과 title과 release_year의 열을 출력
title_year = netflix.loc[[0, 1], ['title', 'release_year']]
title_year.head()
```

```python
# .index : 행 이름 변경
netflix.index = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
# .loc[] : 데이터프레임의 행과 열의 이름을 사용
# B, C행과 title과 release_year의 열을 출력
title_year = netflix.loc[['B', 'C'], ['title', 'release_year']]
title_year.head()
```

```python
# iloc[] : 데이터프레임의 행과 열 번호를 사용
#  3번째, 4번째 행과 1번째, 3번째 열을 출력
type_director = netflix.iloc[[3,4], [1, 3]]
type_director.head()
```

```python
#  0번쩨 행, 0번째 열의 값을 'NO Type'변경
type_director.iloc[0,0] = 'No Type'
type_director.head()
```

```python
# .isin(values) : 각 요소가 values에 포함된 값인지 파악하여 True/False 반환
# type 변수의 값이 TV Show일 경우는 True 반환, 아니라면 False 반환
netflix['type'].isin(['TV Show'])
```

```python
# 논리형 인덱싱 + .isin() : 데이터의 특정 범위만 추출
# type변수의 값이 TV Show일 경우만 추출하여 TV_Show 저장
TV_Show = netflix[netflix['type'].isin(['TV Show'])]
TV_Show.head()
```

```python
# 결측치 확인(3) - .for문
for i in netflix.columns:
    missingValueRate = netflix[i].isna().sum()/len(netflix) *100
    if missingValueRate > 0:
        print("{} null rate: {}%".format(i, round(missingValueRate, 2 )))
```

```python
# 판다스, 넘파이 라이브러리 불러오기
import pandas as pd
import numpy as np
# 파이썬 딕셔너리 정의
customer_dic = {
'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace','Hyemin'],
'comment_length': [150, 200, 50, 300, 120, 180, 75, 160],
'likes': [25, 30, 10, 45, 20, 35, 5, 28],
'is_spam': [False, False, True, np.nan, False, True, np.nan, False],
'has_image': [True, False, True, True, False, False, True, True]
}

# 판다스 데이터프레임 생성
customer = pd.DataFrame(customer_dic)

# 결과 확인
customer
```

```python
# 결측치 처리
customer['is_spam'] = customer['is_spam'].fillna(False)

# 데이터프레임 내용 확인
customer.info()
```

```python
# 그룹화와 집계함수 처리
customer.groupby('has_image')['likes'].mean()
```

```python
# 조건 설정(비교연산자, 논리연산자)
# 댓글 길이 100자 이상, 좋아요 20개 이상, 스팸 댓글이 아닌 글, 이미지가 포함된 글
condition = (
(customer['comment_length'] >= 100) &
(customer['likes'] >= 20) &
(customer['is_spam'] == False) &
(customer['has_image'])
)

# 조건을 만족하는 데이터 필터
winners = customer[condition]

# 결과 확인
winners
```
